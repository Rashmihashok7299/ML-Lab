```
import pandas as pd
 
df_species = pd.read_csv("/naivebayesdataset.csv")
print(df_species)
 
attribute_names = list(df_species.columns)
print("List of Attributes:", attribute_names)
attribute_names.remove('Species')
print("Predicting Attributes:", attribute_names)
table=dict()
priorProb=dict()
 
 
train=df_species.sample(frac=0.8,random_state=100) #random state is a seed value
test=df_species.drop(train.index)
 
print("ddddddddddddddddddddddddddddddd")
print(train)
print("dddddddddddddddddddddddddddddddd")
print(test)
print("dddddddddddddddddddddddd")
 
 
for attr_val, data_subset in train.groupby("Species"):
    from collections import Counter
    valueCount = dict()
    count=0
    for attr_value in attribute_names:
        cnt = Counter(x for x in data_subset[attr_value])
        count=sum(cnt.values())
        valueCount[attr_value]=dict(cnt)
        print("value count", valueCount.values())
        print("counter:-",cnt)
 
    table[attr_val]=valueCount
    priorProb[attr_val]=count
print("*******************************************")
from pprint import pprint
 
print("\n\nThe Resultant table is :\n")
pprint(table)
pprint(priorProb)
 
totalSize=test['Species'].count()
correctPridictions=0
for k, row in test.iterrows():
 
    rowTuple=dict(row)
    print("print row tuple")
    pprint(rowTuple)
    postioriList=list()
    SpeciesList=list()
    for Species in table.keys():
        posteriori = 1.0
        print("RowTuple",rowTuple.keys())
        print("RowValues",rowTuple.values())
        for key in [x for x in rowTuple.keys() if x != 'Species']:
            print(key, "Species:",Species)
            attributeValue=rowTuple.get(key)
            if attributeValue in table[Species][key].keys():
                countList=table[Species][key].values()
                #print("CountList:", countList)
                attributeCount=table[Species][key][attributeValue]
                #print("CountList:",countList)
                #print("SumCountList",sum(countList))
                #print("AttributeCount:",attributeCount)
                #print("key:valuepair",key,":",rowTuple[key])
                posteriori=1.0*attributeCount/sum(countList)*posteriori
 
        posteriori=posteriori*priorProb[Species]
        SpeciesList.append(Species)
        postioriList.append(posteriori)
        print(SpeciesList)
        print(postioriList)
    maxProbInd = postioriList.index(max(postioriList))
    print(rowTuple['Species'], "::::", SpeciesList[maxProbInd])
    if rowTuple['Species'] == SpeciesList[maxProbInd]:
        print(rowTuple['Species'],"::::",SpeciesList[maxProbInd])
        correctPridictions=correctPridictions+1
        print("POSTERIORI OF:",Species,"is:",posteriori)
print("Number of Correct Predictions : Number of Samples",correctPridictions,":",totalSize)
print("Accuracy:",100.0*correctPridictions/totalSize)
```
